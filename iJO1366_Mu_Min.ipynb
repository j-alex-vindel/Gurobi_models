*** Using Multi Scenario from Gurobi by changing the coefficients in the objective function***

```python
import gurobipy as gp
import itertools
from gurobipy import GRB
import pandas as pd
import numpy as np
```

```python
# Read Data

UB = pd.read_csv('iJO1366_UB_n.csv',header=None)[0].tolist()

LB = pd.read_csv('iJO1366_LB_n.csv',header=None)[0].tolist()
 
rxn = pd.read_csv('iJO1366_rxn.csv',header=None)[0].tolist()

met = pd.read_csv('iJO1366_met.csv',header=None)[0].tolist()

b = [0 for i in range(len(met))]

#c = pd.read_csv('iJO1366_c.csv',header=None)[0].tolist()

S = pd.read_csv('iJO1366_S.csv',header=None).values

r = pd.read_csv('iJO1366_r.csv',header=None)[0].tolist() # if reaction i is reversible r[i] = 1

names = pd.read_csv('iJO1366_names.csv',header=None)[0].tolist()
```

```python
bm_core = rxn.index('BIOMASS_Ec_iJO1366_core_53p95M')
def c_coeff(ob):
    for i in range(len(ob)):
        ob[i] = ob[i]*0
    return ob
```

```python
N = [i for i in range(len(met))] # N is for Metabolites
M = [i for i in range(len(rxn))] # M is for Reactions
c = [1 if i == 0 else 0 for i in M] #Coefficient 
m = gp.Model()

# Add Variables

v = m.addVars(M,lb=-GRB.INFINITY,ub=GRB.INFINITY,obj=c, name='v')

m.modelSense = GRB.MINIMIZE
#m.setObjective((1*v[chemical]), GRB.MINIMIZE)

# Add Constr


m.addConstrs((gp.quicksum(S[i,j] * v[j] for j in M) == 0 for i in N), name='FBA')
    
m.addConstrs((LB[j] <= v[j] for j in M), name='LB')

m.addConstrs((v[j] <= UB[j] for j in M), name='UB')


m.numScenarios = len(M)+1

m.Params.ScenarioNumber = 0
m.ScenNName = 'Target %s'%rxn[0]

for i in M:
    if i == 0:
        pass
    else:
        d = [0 for d in range(len(M))]
        d[i] = 1
        e = c_coeff(c)
        m.Params.ScenarioNumber = i+1

        m.ScenNName = 'Target %s'%(rxn[i])
        
        nc = [x+y for (x,y) in zip(e,d)]
#         if i == bm_core:
#             print('***** Biomass *****')
#             print(e[5:i+3])
#             print(nc[5:i+3])
        v[i].ScenNObj = nc[i]*1


m.optimize()
```
